```python
%matplotlib inline
import pandas as pd
import seaborn as sns # great visualization library
```

```python
ad_spend = pd.read_csv(
    'http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', 
    index_col=0)
ad_spend.info()
```

```python
ad_spend.head()
```

### From wide to long format

```python
ad_spend_long = pd.melt(frame=ad_spend, 
        id_vars=['Sales'],
        value_vars=['TV', 'Radio', 'Newspaper'], 
        value_name='Ad Spend', var_name='Channel')
ad_spend_long.head(3)
```

### Does Ad Spend cause Sales?

```python
sns.lmplot(data=ad_spend_long,
           x='Ad Spend', y='Sales',
          col='Channel', hue='Channel');
```
